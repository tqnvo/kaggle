````python
# Import helpful libraries
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split

# Load the data, and separate the target
# Target is the SalePrice
# Load csv file to path
iowa_file_path = '../input/train.csv'
# Read csv file by pandas
home_data = pd.read_csv(iowa_file_path)
#print('All columns of home_data: \n' + str(home_data.columns))
#print('\n')
#print('Head of home_data: \n' + str(home_data.head()))
#print('\n')
#print('Statistics of home_data: \n' + str(home_data.describe()))
#print('\n')
````

````python
# Extract SalePrice as output from csv data
y = home_data.SalePrice
#print('This is statistics of y: \n' + str(y.describe()))
#print('\n')
#print('This is head of y: \n' + str(y.head()))
#print('\n')

# Create X (After completing the exercise, you can return to modify this line!)
# Define 7 original criteria
features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']

features_25 = ['MSSubClass',
                'LotArea',
                'OverallQual',
                'OverallCond',
                'YearBuilt',
                'YearRemodAdd',
                '1stFlrSF',
                '2ndFlrSF',
                'LowQualFinSF',
                'GrLivArea',
                'FullBath',
                'HalfBath',
                'BedroomAbvGr',
                'KitchenAbvGr',
                'TotRmsAbvGrd',
                'Fireplaces',
                'WoodDeckSF',
                'OpenPorchSF',
                'EnclosedPorch',
                '3SsnPorch',
                'ScreenPorch',
                'PoolArea',
                'MiscVal',
                'MoSold',
                'YrSold']

features_critical = [#'MSSubClass',
                'LotArea',
                'OverallQual',
                'OverallCond',
                'YearBuilt',
                'YearRemodAdd',
                '1stFlrSF',
                '2ndFlrSF',
                #'LowQualFinSF',
                'GrLivArea',
                'FullBath',
                #'HalfBath',
                'BedroomAbvGr',
                'KitchenAbvGr',
                'TotRmsAbvGrd',
                'Fireplaces',
                #'WoodDeckSF',
                #'OpenPorchSF',
                #'EnclosedPorch',
                #'3SsnPorch',
                #'ScreenPorch',
                'PoolArea'
                #'MiscVal',
                #'MoSold',
                #'YrSold'
                    ]

features_16 = ['MSSubClass',
                'LotArea',
                'OverallQual',
                'OverallCond',
                'YearBuilt',
                'YearRemodAdd',
                '1stFlrSF',
                '2ndFlrSF',
                #'LowQualFinSF',
                'GrLivArea',
                'FullBath',
                #'HalfBath',
                'BedroomAbvGr',
                'KitchenAbvGr',
                'TotRmsAbvGrd',
                'Fireplaces',
                #'WoodDeckSF',
                #'OpenPorchSF',
                #'EnclosedPorch',
                #'3SsnPorch',
                #'ScreenPorch',
                'PoolArea',
                #'MiscVal',
                #'MoSold',
                'YrSold'
                    ]


# Select columns corresponding to features, and preview the data
# Extract the data based on defined criteria
X = home_data[features]
X.head()

X_25 = home_data[features_25]
X_25.head()

X_critical = home_data[features_critical]
X_critical.head()
X_critical.describe()

X_16 = home_data[features_16]
X_16.head()
X_16.describe()

#X.describe()
#print('All columns of X: \n' + str(X.columns))
#print('\n')
#print('This is statistics of X: \n' + str(X.describe()))
#print('\n')
#print('This is head of X: \n' + str(X.head()))
#print('\n')
````

````python
# Split into validation and training data
# The big set of both X and y are now split into 2 sets
# training set and test set
# test set is also called validation set
# X --> train_X and val_X
# y --> train_y and val_y

# Train test sets based on X
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)

# Train test sets based on X_25
train_X_25, val_X_25, train_y_25, val_y_25 = train_test_split(X_25, y, random_state=1)

# Train test sets based on X_critical
train_X_critical, val_X_critical, train_y_critical, val_y_critical = train_test_split(X_critical, y, random_state=1)

# Train test sets based on X_16
train_X_16, val_X_16, train_y_16, val_y_16 = train_test_split(X_16, y, random_state=1)

# Define a random forest model
rf_model = RandomForestRegressor(random_state=1)
rf_model_25 = RandomForestRegressor(random_state=1)
rf_model_critical = RandomForestRegressor(random_state=1)
rf_model_16 = RandomForestRegressor(random_state=1)

# Trying to fit train_X as input with train_y as output
rf_model.fit(train_X, train_y)
rf_model_25.fit(train_X_25, train_y_25)
rf_model_critical.fit(train_X_critical, train_y_critical)
rf_model_16.fit(train_X_16, train_y_16)

# After fitting we will obtain weight w and bias b
# Weight w and bias b will be used for prediction
# The prediction is performed by using val_X as input
# The prediction output will be called rf_val_prediction

# Computing rf_val_predicitons based on val_X
rf_val_predictions = rf_model.predict(val_X)
rf_val_predictions_25 = rf_model_25.predict(val_X_25)
rf_val_predictions_critical = rf_model_critical.predict(val_X_critical)
rf_val_predictions_16 = rf_model_16.predict(val_X_16)

# Computing mean absolute error between val_y and rf_val_predictions
rf_val_mae = mean_absolute_error(rf_val_predictions, val_y)
rf_val_mae_25 = mean_absolute_error(rf_val_predictions_25, val_y_25)
rf_val_mae_critical = mean_absolute_error(rf_val_predictions_critical, val_y_critical)
rf_val_mae_16 = mean_absolute_error(rf_val_predictions_16, val_y_16)

print("Validation MAE for Random Forest Model ori : {:,.0f}".format(rf_val_mae))
print("Validation MAE for Random Forest Model 25  : {:,.0f}".format(rf_val_mae_25))
print("Validation MAE for Random Forest Model cri  : {:,.0f}".format(rf_val_mae_critical))
print("Validation MAE for Random Forest Model 16  : {:,.0f}".format(rf_val_mae_16))
````
